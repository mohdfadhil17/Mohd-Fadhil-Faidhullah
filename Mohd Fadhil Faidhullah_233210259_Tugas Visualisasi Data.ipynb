```python
import matplotlib.pyplot as plt

Usia = ['20', '21', '22', '23', '24']
Gender = [4,4,3,3,3]

#grafiknya
plt.figure(figsize=(4, 4))
plt.pie(Gender, labels=Usia, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Jumlah Gender Per Usia')
plt.axis('equal') 
plt.show()
```

Pie chart ini berisi jumlah gender per usia. Jumlah jumlah gender tiap usia adalah usia 20 = 4, usia 21 = 4, usia 22 = 3, usia 23 = 3, usia 24 = 3, setelah coding pie chart dibuat, didapatkan pie chart diatas ini.

```python
import matplotlib.pyplot as plt

Usia = ['20', '21', '22', '23', '24']
Jumlah_lulus = [6,4,7,3,2]

#grafiknya
plt.figure(figsize=(8, 6))
plt.plot(Usia, Jumlah_lulus, marker='+', color='r', linestyle='-') 
plt.title('Line Chart Jumlah Lulus per Usia')
plt.xlabel('usia')
plt.ylabel('Jumlah lulus')
plt.grid(True)
plt.show()
```

line chart ini berisi jumlah kelulusan tiap usia, jumlah yang lulus tiap usia adalah usia 20 = 6 orang, 21 = 4 orang, 22 = 7 orang, 23 = 3 orang, dan 24 = 2 orang. Setelah coding line chart dibuat, didapatkan line chart seperti diatas ini.

```python
import matplotlib.pyplot as plt

Usia = ['20', '21', '22', '23', '24']
Jumlah_lulus = [6,4,7,3,2]

#grafiknya
plt.figure(figsize=(8, 4))
plt.bar(Usia, Jumlah_lulus, color='b') 
plt.title('bar Chart Jumlah Lulus per Usia')
plt.xlabel('hari')
plt.ylabel('Jumlah Penjualan')
plt.grid(False)
plt.show()
```

Bar chart ini berisi jumlah kelulusan tiap usia, jumlah yang lulus tiap usia adalah usia 20 = 6 orang, 21 = 4 orang, 22 = 7 orang, 23 = 3 orang, dan 24 = 2 orang. Setelah coding line chart dibuat, didapatkan bar chart seperti diatas ini.
